```python
import pandas as pd
import numpy as np
```

```python
farm_ads = pd.read_csv('farm-ads', sep = " ", error_bad_lines = False, header = None)
farm_vect = pd.read_csv('farm-ads-vect',sep = " ", error_bad_lines = False, header = None)
```

```python
df = farm_ads.merge(farm_vect, right_index = True, left_index = True)
```

```python
df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

for i in df.columns:
    if df[i].dtypes == "O":
        print(i)
        df[i] = df[i].astype(str)
        label_encoder.fit(df[i])
        df[i] = label_encoder.transform(df[i])
        
# Resources used:
# https://www.geeksforgeeks.org/ml-label-encoding-of-datasets-in-python/#
```

```python
df['Target'] = df['0_x'] + df['0_y']
```

```python
df = df.drop(['0_x', '0_y'], axis = 1)
```

```python
df
```

```python
# Please note, kaggle link below taught me how to clean this dataset.

X = df.drop('Target', axis = 1)
y = df['Target']

# Resources used:
# https://www.kaggle.com/code/bcantt/farm-ads-starter-ml
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
df = sc.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# According to the resource below, PCA, LDA and ICA are dimension reduction algorithms.

from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import FastICA

# Resources used:
# https://towardsdatascience.com/pca-lda-ica-a-components-analysis-algorithms-comparison-c5762c4148ff
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html
# https://scikit-learn.org/stable/modules/generated/sklearn.discriminant_analysis.LinearDiscriminantAnalysis.html
# https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.FastICA.html
```

```python
pca = PCA()
lda = LinearDiscriminantAnalysis()
ica = FastICA(max_iter = 1000)
```

```python
X_reduced_pca = pca.fit_transform(X_train)
X_reduced_lda = lda.fit_transform(X_train, y_train)
X_reduced_ica = ica.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
```

```python
pca_scores = []
lda_scores = []
ica_scores = []
```

```python
for n_components in range(1, X_train.shape[1] + 1):
    
    X_reduced_pca = pca.fit_transform(X_train)[:, :n_components]
    pca_scores.append(classifier.fit(X_reduced_pca, y_train).score(X_reduced_pca, y_train))
    
    X_reduced_lda = lda.fit_transform(X_train, y_train)[:, :n_components]
    lda_scores.append(classifier.fit(X_reduced_lda, y_train).score(X_reduced_lda, y_train))
    
    X_reduced_ica = ica.fit_transform(X_train)[:, :n_components]
    ica_scores.append(classifier.fit(X_reduced_ica, y_train).score(X_reduced_ica, y_train))
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, X_train.shape[1] + 1), pca_scores, label = "PCA")
plt.plot(range(1, X_train.shape[1] + 1), lda_scores, label = "LDA")
plt.plot(range(1, X_train.shape[1] + 1), ica_scores, label = "ICA")
plt.xlabel("Components")
plt.ylabel("Accuracy Score")
plt.legend()
plt.show()
```
