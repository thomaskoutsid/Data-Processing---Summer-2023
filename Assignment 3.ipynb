```python
averages = csv_df.groupby('Team')[['Salary', 'Bonus %']].mean()
print(averages)
```

```python
def top(dataframe):
    return dataframe.nlargest(5, 'Salary')

paid = top(csv_df)
print("Five Highest Paid:\n", paid)
```

```python
genders = csv_df.groupby('Gender')

for i, j in genders:
    print(f"Five Highest Paid {i}:")
    genders_paid = top(j)
    print(genders_paid, "\n")
```
